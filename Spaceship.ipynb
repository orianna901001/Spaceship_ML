<h3>train.csv的資料集介紹</h3>

約8700條乘客的個人紀錄，用於訓練模型。
<ol>
<li>PassengerId:每位乘客的ID</li>
<li>HomePlanet:乘客離開的行星，通常是他们的永久居住行星</li>
<li>CryoSleep: 指示乘客是否在航行期間被選擇進入假死狀態。處於冷凍睡眠狀態的乘客被限制在他們的機艙內</li>
<li>Cabin: 乘客居住的機艙號。採用採取”deck/num/side”即”甲板/編號/側面”的形式，其中側面可以是P代表左舷，S代表右舷。</li>
<li>Destination: 乘客將要下船的星球</li>
<li>Age:乘客的年龄</li>
<li>VIP: 乘客在航行期間是否支付了特殊VIP服務的費用</li>
<li>RoomService, FoodCourt, ShoppingMall, Spa, VRDeck:乘客在泰坦尼克號宇宙飛船的許多豪華設施中收取的費用</li>
<li>Name:乘客的名字和姓氏</li>
<li>Transported:乘客是否被送到另一個維度。</li>

結果/推測：在資料前處理時，Cabin這個欄位在字串中是這樣表示的：(F/5/P)，所以在做int轉換決定時，猶豫了很久是否只要留第一個F，然後丟棄其他的字元，這讓我突然思考到跟啤酒尿布經典題的一個共通點，這些看似不相關的資料，可能被我隨手丟棄的話，會不會影響到最後Accuracy的數據，所以我推測或許能將後面的數據留下，不破壞完整train的資料，能稍微提高Accuracy的數據；後來再回去實驗過後，透過sns.pairplot得知剩下的欄位與結果的預測相關性不高，故沒有保留此兩個欄位
